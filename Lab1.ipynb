```python
import random

#a)
def simulare_experiment():
    red, blue, black = 3, 4, 2
    zar = random.randint(1, 6)
    prime = {2, 3, 5}

    if zar in prime:
        black += 1
    elif zar == 6:
        red += 1
    else:
        blue += 1

    urna = ['red'] * red + ['blue'] * blue + ['black'] * black
    bila = random.choice(urna)
    return bila == 'red'

#b)
N = 200000
rezultate = [simulare_experiment() for _ in range(N)]
prob_simulata = sum(rezultate)/N
print(f"probabilitate simulata de a extrage o bilă rosie: {prob_simulata:.6f}")

#c)
#luam cele 3 cazuri:
#1.cand zarul este prim, urna are 3 rosii din 10 bile
#P(rosie∣prim)=3/10
#2.cand zarul este 6, urna are 4 rosii din 10 bile
#P(rosie∣6)=4/10
#3.csnd zarul este 1 sau 4, urna are din nou 3 rosii din 10 bile, deci
#P(rosie∣1,4)=3/10
#probabilitate total:
#P(rosie)=P(prim)⋅P(rosie∣prim)+P(6)⋅P(rossie∣6)+P(1,4)⋅P(rosie∣1,4)
#prob_teoretica = (3/6)*(3/10) + (1/6)*(4/10) + (2/6)*(3/10)
print(f"probabilitate teoretica: 0.316667")


if prob_simulata > prob_teoretica:
    print(f"probabilitatea simulata este mai mare decat cea teoretica cu {prob_simulata-prob_teoretica:.6f}.")
elif prob_simulata < prob_teoretica:
    print(f"probabilitatea simulata este mai mica decat cea teoretica cu { prob_teoretica-prob_simulata:.6f}.")
else:
    print("probabilitaile sunt la fel")
```
